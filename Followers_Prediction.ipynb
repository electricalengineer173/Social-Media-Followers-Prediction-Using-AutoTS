```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = pd.read_csv("Book1.csv")
data.drop(data.tail(1).index, inplace=True)
data.head()
```

```python
plt.figure(figsize=(15, 10))
#sns.set_theme(style="whitegrid")
plt.title("Monthly Followers Gained")
sns.barplot(x="period_end", y="followers_gained", data=data,color='green')
plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.set_theme(style="whitegrid")
plt.title("Monthly Net Followers")
sns.barplot(x="followers_total", y="period_end", data=data)
plt.show()
```

```python
from autots import AutoTS
# Automate Time Series  Auto-TS is a part of AutoML that will automate the time series forecasting process.

#Some of the features of AutoTS are:

#Around 20 pre-defined models like ARIMA, ETS, VECM, etc. are available to use with thousands of possible hyperparameters.
#It works on finding the best model according to the seasonality and trend of the data by genetic programming
#AutoTS can handle both Univariate and Multi-Variate Datasets.
#AutoTS itself clears the data from any NaN columns or outliers which are present.
```

```python
model = AutoTS(forecast_length=1, frequency='infer', ensemble='simple')
```

```python
model = model.fit(data, date_col='period_end', value_col='followers_gained', id_col=None)
prediction = model.predict()
forecast = prediction.forecast
#print(forecast)
```
